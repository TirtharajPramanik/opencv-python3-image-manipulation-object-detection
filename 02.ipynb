```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
nzb_bgr = cv2.imread('assets/New_Zealand_Boat.jpg', 1)
nzb_rgb = cv2.cvtColor(nzb_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(nzb_rgb)
```

```python
nzb_cr = nzb_rgb[200:400,300:600]

plt.imshow(nzb_cr)
```

```python
nzb_cr_rz = cv2.resize(nzb_cr, None,fx=2,fy=2)

plt.imshow(nzb_cr_rz)
```

```python
nzb_cr_rz1 = cv2.resize(nzb_cr, (100,200), interpolation=cv2.INTER_AREA)

plt.imshow(nzb_cr_rz1)
```

```python
dw = 100
ar = dw/nzb_cr.shape[1]
dh = int(ar*nzb_cr.shape[0])

nzb_cr_rz2 = cv2.resize(nzb_cr, (dw,dh), interpolation=cv2.INTER_AREA)

plt.imshow(nzb_cr_rz2)
```

```python
nzb_cr_rz_fl = cv2.flip(nzb_cr_rz, 1)

cv2.imwrite('assets/New_Zealand_Boat_modified.png', nzb_cr_rz_fl[:,:,::-1])

Image('assets/New_Zealand_Boat_modified.png')
```
