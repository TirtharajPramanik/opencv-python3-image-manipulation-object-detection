```python
nzl_img_bgr = cv.imread('assets/New_Zealand_Lake.jpg',cv.IMREAD_COLOR)
plt.imshow(nzl_img_bgr)
```

```python
b,g,r = cv.split(nzl_img_bgr)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r,cmap='gray');plt.title('Red Chennel');
plt.subplot(142)
plt.imshow(g, cmap='gray')
plt.title('Green Chennel')
plt.subplot(143)
plt.imshow(b, cmap='gray')
plt.title('Blue Chennel')

nzl_mrg_img_bgr = cv.merge((b,g,r))
plt.subplot(144);plt.imshow(nzl_mrg_img_bgr[:,:,::-1]);plt.title('Merged Chennel');
```

```python
nzl_img_rgb = cv.cvtColor(nzl_img_bgr, cv.COLOR_BGR2RGB)
plt.imshow(nzl_img_rgb)
```

```python
nzl_img_hsv = cv.cvtColor(nzl_img_bgr, cv.COLOR_BGR2HSV)
h,s,v=cv.split(nzl_img_hsv)

plt.figure(figsize=[20, 5])
plt.subplot(141);plt.imshow(h, cmap='gray');plt.title('Hue Chennel')
plt.subplot(142);plt.imshow(s, cmap='gray');plt.title('Saturation Chennel')
plt.subplot(143);plt.imshow(v, cmap='gray');plt.title('Value Chennel')
plt.subplot(144);plt.imshow(nzl_img_rgb);plt.title('Original')
```

```python
new_h = h+10
nzl_img_modified_hsv = cv.merge((new_h,s,v))
nzl_img_modified_rgb = cv.cvtColor(nzl_img_modified_hsv, cv.COLOR_HSV2RGB)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h);plt.title('H')
plt.subplot(142);plt.imshow(new_h);plt.title('New H')
plt.subplot(143);plt.imshow(nzl_img_rgb);plt.title('RGB')
plt.subplot(144);plt.imshow(nzl_img_modified_rgb);plt.title('New HSV')
```

```python
nzl_img_modified_bgr = cv.cvtColor(nzl_img_modified_hsv, cv.COLOR_HSV2BGR)

cv.imwrite('assets/New_Zealand_Lake_modified.png', nzl_img_modified_bgr)
Image(filename='assets/New_Zealand_Lake_modified.png')
```
