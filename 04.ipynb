```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
nzc_bgr = cv2.imread('assets/New_Zealand_Coast.jpg', 1)
nzc = cv2.imread('assets/New_Zealand_Coast.jpg', 1)

plt.imshow(nzc)
```

```python
mat = np.ones(nzc.shape, 'uint8') *50

nzc_br = cv2.add(nzc,mat)
nzc_dr = cv2.subtract(nzc,mat)

plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(nzc_dr);plt.title('Darker')
plt.subplot(132);plt.imshow(nzc);plt.title('Original')
plt.subplot(133);plt.imshow(nzc_br);plt.title('Brighter')
```

```python
mat1 = np.ones(nzc.shape) * .8
mat2 = np.ones(nzc.shape) * 1.2

nzc_cn_dr = np.uint8(cv2.multiply(np.float64(nzc), mat1))
nzc_cn_br = np.uint8(cv2.multiply(np.float64(nzc), mat2))

plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(nzc_cn_dr);plt.title('Lower Contrast')
plt.subplot(132);plt.imshow(nzc);plt.title('Original')
plt.subplot(133);plt.imshow(nzc_cn_br);plt.title('Higher Contrast')


nzc_cn_dr = np.uint8(np.clip(cv2.multiply(np.float64(nzc), mat1), 0, 255))
nzc_cn_br = np.uint8(np.clip(cv2.multiply(np.float64(nzc), mat2), 0, 255))

plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(nzc_cn_dr);plt.title('Lower Contrast')
plt.subplot(132);plt.imshow(nzc);plt.title('Original')
plt.subplot(133);plt.imshow(nzc_cn_br);plt.title('Higher Contrast')
```

```python
wd = cv2.imread('assets/building-windows.jpg',0)
_, wd_thresh = cv2.threshold(wd, 100, 255, cv2.THRESH_BINARY)

plt.figure(figsize=[18,5])
plt.subplot(121);plt.imshow(wd,'gray');plt.title('Original')
plt.subplot(122);plt.imshow(wd_thresh,'gray');plt.title('Thresholded')

print(wd_thresh.shape)
```

```python
mc = cv2.imread('assets/Piano_Sheet_Music.png',0)

_, mc_thresh = cv2.threshold(mc, 50,255,cv2.THRESH_BINARY)
_, mc_thresh1 = cv2.threshold(mc, 130,255,cv2.THRESH_BINARY)
mc_thresh_adp = cv2.adaptiveThreshold(mc, 255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,7)

plt.figure(figsize=[18,15])
plt.subplot(221);plt.imshow(mc,'gray');plt.title('Original')
plt.subplot(222);plt.imshow(mc_thresh,'gray');plt.title('Low Threshold')
plt.subplot(223);plt.imshow(mc_thresh1,'gray');plt.title('High Threshold')
plt.subplot(224);plt.imshow(mc_thresh_adp,'gray');plt.title('Adaptive Threshold')
```
