```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
img = cv2.imread('assets/Apollo_11_Launch.jpg', 1)

plt.imshow(img[:,:,::-1])
```

```python
# img_ln = img.copy()

# cv2.line(img_ln, (200,100),(400,100),(0,255,255),thickness=5, lineType=cv2.LINE_AA)
cv2.line(img, (200,100),(400,100),(0,255,255),thickness=5, lineType=cv2.LINE_AA)

# plt.imshow(img_ln[:,:,::-1])
plt.imshow(img[:,:,::-1])
```

```python
# img_cc = img.copy()

# cv2.circle(img_cc, (900,500), 100,(0,0,255),5,lineType=cv2.LINE_AA)
cv2.circle(img, (900,500), 100,(0,0,255),5,lineType=cv2.LINE_AA)

# plt.imshow(img_cc[:,:,::-1])
plt.imshow(img[:,:,::-1])
```

```python
# img_rc = img.copy()

# cv2.rectangle(img_rc, (500,100),(700,600), (255,0,255),5,lineType=cv2.LINE_8)
cv2.rectangle(img, (500,100),(700,600), (255,0,255),5,lineType=cv2.LINE_8)

# plt.imshow(img_rc[:,:,::-1])
plt.imshow(img[:,:,::-1])
```

```python
# img_tx = img.copy()

text = "Apollo 11 Saturn V Launch, July 16, 1969"
# cv2.putText(img_tx,text,(200,700),cv2.FONT_HERSHEY_PLAIN,2.3,(0,255,0),2,cv2.LINE_AA)
cv2.putText(img,text,(200,700),cv2.FONT_HERSHEY_PLAIN,2.3,(0,255,0),2,cv2.LINE_AA)

# plt.imshow(img_tx[:,:,::-1])
plt.imshow(img[:,:,::-1])
```

```python
cv2.imwrite('assets/Apollo_11_Launch_modified.png', img)

Image('assets/Apollo_11_Launch_modified.png')
```
